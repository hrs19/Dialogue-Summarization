```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data preparation
data = {
    'Model': ['BART-base', 'BART-large-cnn', 'FLAN-T5-small', 'FLAN-T5-base'],
    'Fine-Tuning Strategy': ['Encoder Only', 'LoRA', 'Full Fine-tune', 'LoRA'],
    'ROUGE-L': [0.4159, 0.3053, 0.4020, 0.4284]
}

df = pd.DataFrame(data)

# Create the plot
plt.figure(figsize=(10, 6))
chart = sns.barplot(x='Model', y='ROUGE-L', hue='Fine-Tuning Strategy', data=df, palette='viridis')

# Customize the plot
plt.title('ROUGE-L Scores by Model and Fine-Tuning Strategy', fontsize=14)
plt.xlabel('Model', fontsize=12)
plt.ylabel('ROUGE-L Score', fontsize=12)
# plt.ylim(0.25, 0.45)  # Set limits for y-axis to make differences more visible
plt.legend(title='Fine-Tuning Strategy', fontsize=10, title_fontsize='13')

# Improve layout
plt.tight_layout()

# Save the plot to a file
plt.savefig('rouge_l_scores.pdf', format='pdf', dpi=300)

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data preparation
data = {
    'Model and Strategy': [
        'BART-base (Encoder Only)', 
        'BART-large-cnn (LoRA)', 
        'FLAN-T5-small (Full Fine-tune)', 
        'FLAN-T5-base (LoRA)'
    ],
    'ROUGE-L': [0.4159, 0.3053, 0.4020, 0.4284]
}

df = pd.DataFrame(data)

# Create the plot
plt.figure(figsize=(10, 6))
bar_chart = sns.barplot(x='Model and Strategy', y='ROUGE-L', data=df, palette='deep')

# Customize the plot
plt.title('ROUGE-L Scores by Model and Fine-Tuning Strategy', fontsize=16)
plt.xlabel('Model and Fine-Tuning Strategy', fontsize=14)
plt.ylabel('ROUGE-L Score', fontsize=14)
# plt.xticks(rotation=45)  # Rotate labels to avoid overlap
# plt.ylim(0.25, 0.45)  # Set limits for y-axis to make differences more visible

# Improve layout
plt.tight_layout()

# Save the plot to a file
plt.savefig('rouge_l_scores.pdf', format='pdf', dpi=300)

# Show the plot
plt.show()
```
